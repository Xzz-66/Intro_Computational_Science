# Phase-space Plotting

## Matplotlib function: streamplot

### MoMo's Practice


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters 
g = 9.80
L = 1.0

# Grid for theta (position) and omega (velocity)
theta = np.linspace(-2 * np.pi, 2 * np.pi, 100)
omega = np.linspace(-10, 10, 100)
THETA, OMEGA = np.meshgrid(theta, omega)

# define the ODEs
dtheta_dt = OMEGA
domega_dt = -g*np.sin(THETA) / L

# plot Phase-Space using streamplot
plt.figure(figsize=(10,6))
plt.streamplot(THETA, OMEGA, dtheta_dt, domega_dt, density=1.2, color='grey')
plt.title('Phase Space of a Simple Pendulum')
plt.xlabel(r'$\theta$ (rad)')
plt.ylabel(r'$\omega = \dot{\theta}$ (rad/s)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
